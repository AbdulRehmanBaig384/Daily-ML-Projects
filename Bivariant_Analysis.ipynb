```python
sns.boxplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Survived'])
plt.show()
```

```python
sns.displot(x=titanic[titanic['Survived']==0]['Age'])
sns.displot(x=titanic[titanic['Survived']==1]['Age'])
```

```python
sns.displot(data=titanic,x='Survived',kind='kde',fill=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# KDE plot without histogram
sns.displot(data=titanic, x="Age", hue="Survived", kind="kde", fill=False)

plt.title("Age Distribution by Survival (KDE)")
plt.show()
```

```python
# If we want to analayze the graph between categorial and categorial so we should adopt heat map
print(pd.crosstab(titanic['Pclass'],titanic['Survived']))

sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```
