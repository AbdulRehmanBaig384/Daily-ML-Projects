```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv')
print(df)
```

```python
df=df.iloc[:,1:]
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])  #c=df['placement']is use to determine which candidate got placement
```

```python
X=df.iloc[:,:2]
Y=df.iloc[:,-1:]
print(Y)
```

```python
from sklearn.model_selection import train_test_split
 # train_test_split(X,Y,test_size=0.1)  #test_size is used to indicate how much data in percentage must be test 0.1 means 10 
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
X_train=scalar.fit_transform(X_train)
print(X_train)
```

```python
X_test=scalar.transform(X_test)
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
ab=LogisticRegression()
```

```python
# Model Training
ab.fit(X_train,Y_train)
```

```python
y_pred=ab.predict(X_test)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
